```python
import matplotlib
matplotlib.use('Agg')

import pandas as pd
import matplotlib.pyplot as plt
import io

data = [
    {"Duration": 50, "Pulse": 110, "Maxpulse": 130, "Calories": 400},
    {"Duration": 60, "Pulse": 70,  "Maxpulse": 120, "Calories": 500}
]

df = pd.DataFrame(data)

df.plot()

buffer = io.BytesIO()
plt.savefig(buffer, format='png')
buffer.seek(0)

from IPython.display import display, Image
display(Image(buffer.read()))
```
