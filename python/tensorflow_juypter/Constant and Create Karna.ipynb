```python
import numpy as np
import tensorflow as tf

# NumPy array create karna
arr = np.array([1, 2, 3, 4, 5])

print("NumPy Array:")
print(arr)

# TensorFlow tensor me convert karna
tensor = tf.constant(arr)

print("\nTensorFlow Tensor:")
print(tensor)

# TensorFlow se calculation
result = tensor * 2

print("\nTensorFlow Result (array * 2):")
print(result)
```
