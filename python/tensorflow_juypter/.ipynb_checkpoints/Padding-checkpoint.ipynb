```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

sequences = [
    [1, 2, 3, 4],
    [1, 2, 3, 5],
    [1, 2, 7, 8, 9, 10]
]

padded = pad_sequences(sequences)

print(padded) # 0 will add at first location
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

sequences = [
    [1, 2, 3, 4],
    [1, 2, 3, 5],
    [1, 2, 7, 8, 9, 10]
]

padded = pad_sequences(sequences, padding="post") # 0 will add at first location

print(padded) #0 will add at last location
```
