```python
import numpy as np
import matplotlib.pyplot as plt

# 10x10 RGB image create karna
image = np.zeros((10, 10, 3), dtype=np.uint8)

# Kuch pixels ko color dena
image[:5, :5] = [255, 0, 0]      # Red block
image[:5, 5:] = [0, 255, 0]      # Green block
image[5:, :5] = [0, 0, 255]      # Blue block
image[5:, 5:] = [255, 255, 0]    # Yellow block

# Image show karna
plt.imshow(image)
plt.title("Generated 10x10 Image")
plt.axis("off")
plt.show()

# Complete pixel matrix print karna
print("Pixel Matrix Shape:", image.shape)
print("\nPixel Values:\n")
print(image)

# Example: ek specific pixel ki RGB value
row, col = 2, 2
print(f"\nPixel ({row},{col}) RGB Value:")
print(image[row, col])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 16x16 image, default white background
img = np.ones((16, 16, 3), dtype=np.uint8) * 255

# Colors
GREEN = [0, 180, 0]
BROWN = [139, 69, 19]

# Tree leaves
img[2:4, 6:10] = GREEN
img[4:6, 5:11] = GREEN
img[6:8, 4:12] = GREEN
img[8:10, 5:11] = GREEN

# Tree trunk
img[10:14, 7:9] = BROWN

# Display image (zoomed)
plt.figure(figsize=(6, 6))
plt.imshow(img, interpolation="nearest")
plt.title("Pixel Tree")
plt.grid(True)
plt.show()

# Pixel matrix
print("Image Shape:", img.shape)
print("\nPixel Data:")
print(img)
```
