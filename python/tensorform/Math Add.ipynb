tf.add() TensorFlow me do tensors ko element-wise add karne ke liye use hota hai.
Ye x + y jaisa hi kaam karta hai, aur agar shapes compatible hon to broadcasting bhi follow karta hai.

```python
import tensorflow as tf
```

```python
a = 10
b = 5
tf.add(a,b)
```

```python
arr = [1,2,3,4]
arr2 = [5,6,7,8]
tf.add(arr, arr2)
```

```python
tupleval = (1,2,3,4)
tupleval2 = (5,6,7,8)
tf.add(tupleval, tupleval2)
```

```python
tf.add(arr, tupleval2)
```

```python
tf.add(tupleval, b)
```

```python
c = [10,15]
tf.add(tupleval, c)
```
