```python
from tensorflow.keras.preprocessing.text import Tokenizer

sentences = [
    "I love my parents",
    "I love my garden",
    "I love my plants",
    "I love to visit different places"
]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

print(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(sentences)

print(sequences)
```
