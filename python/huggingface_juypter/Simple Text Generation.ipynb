```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained("gpt2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# Input text
input_text = "I love"

# Tokenize
inputs = tokenizer(input_text, return_tensors="pt")

# Generate next words
outputs = model.generate(
    **inputs,
    max_new_tokens=5,         # Generate 5 new tokens
    temperature=0.7,          # Lower = focused, Higher = creative [web:39]
    do_sample=True,           # Use sampling for variety
    top_k=10,                 # Sample from top 10 tokens
    num_return_sequences=3    # Generate 3 different sequences
)

# Decode and print results
print(f"Input: {input_text}")
print("-" * 50)
for i, output in enumerate(outputs, 1):
    generated_text = tokenizer.decode(output, skip_special_tokens=True)
    print(f"Output {i}: {generated_text}")

'''
Input: I love
--------------------------------------------------
Output 1: I love programming and I love code
Output 2: I love spending time with my family
Output 3: I love to travel and explore new places
'''
```
