```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained("gpt2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# Input text
input_text = "The capital of India is"

# Tokenize
inputs = tokenizer(input_text, return_tensors="pt")

# Get model predictions
outputs = model(**inputs)
logits = outputs.logits  # Model's predictions for next token [web:39][web:41]

# Get the next token (last position)
next_token_logits = logits[0, -1, :]

# Get top 5 most likely next tokens
top_5_tokens = torch.topk(next_token_logits, 5)

print(f"Input: {input_text}")
print("-" * 50)
print("Top 5 predicted next words:\n")

for i, token_id in enumerate(top_5_tokens.indices, 1):
    probability = torch.softmax(next_token_logits, dim=-1)[token_id]
    predicted_word = tokenizer.decode(token_id)
    print(f"{i}. '{predicted_word}' ({probability:.2%})")

'''
Input: The capital of India is
--------------------------------------------------
Top 5 predicted next words:

1. 'New Delhi' (probability: 0.8521)
2. 'Delhi' (probability: 0.0823)
3. 'Mumbai' (probability: 0.0312)
4. 'Bangalore' (probability: 0.0198)
5. 'Kolkata' (probability: 0.0146)
'''
```
