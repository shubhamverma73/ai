```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(
    "bert-base-uncased"
)

tokens = tokenizer("I love AI")

print(tokens)
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Token IDs ko original text mein reverse karo
print(tokenizer.decode([1045, 2293, 9932]))
# Output: "i love ai"

# Ya individual tokens dekho
print(tokenizer.convert_ids_to_tokens([1045, 2293, 9932]))
# Output: 'i', 'love', 'ai']
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(
    "bert-base-uncased"
)

tokens = tokenizer("I love Bhadohi")

print(tokens)
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Token IDs ko original text mein reverse karo
print(tokenizer.decode([1045, 2293, 1038, 16102, 11631, 2072]))
# Output: "i love bhadohi"

# Ya individual tokens dekho
print(tokenizer.convert_ids_to_tokens([1045, 2293, 1038, 16102, 11631, 2072]))
# Output: ['i', 'love', 'b', 'had', 'oh', 'i']
```
