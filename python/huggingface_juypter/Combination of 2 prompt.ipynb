```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# 1. Load model and tokenizer
print("Loading GPT-2 model...")
model = AutoModelForCausalLM.from_pretrained("gpt2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# 2. Set device (CPU or GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f"Using device: {device}")

# 3. Input text
input_text = "Artificial intelligence is"

# 4. Tokenize input
inputs = tokenizer(input_text, return_tensors="pt").to(device)

# 5. Generate text
print(f"\nInput: {input_text}")
print("-" * 50)

outputs = model.generate(
    **inputs,
    max_new_tokens=10,
    temperature=0.8,
    do_sample=True,
    top_k=20,
    num_return_sequences=3
)

# 6. Decode and display results
print("Generated outputs:\n")
for i, output in enumerate(outputs, 1):
    generated_text = tokenizer.decode(output, skip_special_tokens=True)
    print(f"{i}. {generated_text}")

# 7. Next word prediction
print("\n" + "=" * 50)
print("NEXT WORD PREDICTION")
print("=" * 50)

input_text2 = "Python is a programming"
inputs2 = tokenizer(input_text2, return_tensors="pt").to(device)

outputs2 = model(**inputs2)
logits2 = outputs2.logits
next_token_logits2 = logits2[0, -1, :]

# Get top 5 tokens
top_5 = torch.topk(next_token_logits2, 5)

print(f"\nInput: {input_text2}")
print("-" * 50)
print("Top 5 predicted next words:\n")

for i, token_id in enumerate(top_5.indices, 1):
    probability = torch.softmax(next_token_logits2, dim=-1)[token_id]
    predicted_word = tokenizer.decode(token_id)
    print(f"{i}. '{predicted_word}' ({probability:.2%})")
```
