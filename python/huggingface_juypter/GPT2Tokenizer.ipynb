```python
from transformers import GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Sahi tareeka
print(tokenizer.encode("I love AI"))
# Output: [1045, 3186, 17256]

# Ya hasher
tokens = tokenizer("I love AI")
print(tokens['input_ids'])
# Output: [1045, 3186, 17256]
```
