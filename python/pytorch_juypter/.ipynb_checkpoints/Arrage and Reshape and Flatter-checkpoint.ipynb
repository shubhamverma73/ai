```python
import torch

x = torch.arange(12)

print("Original")
print(x)
print(x.shape)

x = x.reshape(3,4)

print("\nReshape")
print(x)
print(x.shape)

x = x.flatten()

print("\nFlatten")
print(x)
print(x.shape)
```
